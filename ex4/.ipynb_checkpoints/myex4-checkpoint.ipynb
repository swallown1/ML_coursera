```python
def getDatumImg(row):
    """
    Function that is handed a single np array with shape 1x400,
    crates an image object from it, and returns it
    """
    width, height = 20, 20
    square = row[1:].reshape(width,height)
    return square.T
    
def displayData(indices_to_display = None):
    """
    Function that picks 100 random rows from X, creates a 20x20 image from each,
    then stitches them together into a 10x10 grid of images, and shows it.
    """
    width, height = 20, 20
    nrows, ncols = 10, 10
    if not indices_to_display:
        indices_to_display = random.sample(range(X.shape[0]), nrows*ncols)
        
    big_picture = np.zeros((height*nrows,width*ncols))
    
    irow, icol = 0, 0
    for idx in indices_to_display:
        if icol == ncols:
            irow += 1
            icol  = 0
        iimg = getDatumImg(X[idx])
        big_picture[irow*height:irow*height+iimg.shape[0],icol*width:icol*width+iimg.shape[1]] = iimg
        icol += 1
    fig = plt.figure(figsize=(6,6))
    img = scipy.misc.toimage( big_picture )
    plt.imshow(img,cmap = cm.BrBG_r)
```

# 1.2.3 展开参数

### 当我们使用高级优化方法来优化神经网络时，我们需要将多个参数矩阵展开，才能传入优化函数，然后再恢复形状。

```python
def serialize(t1,t2):
    return np.r_[t1.flatten(),t2.flatten()]
```

```python
thetas = serialize(theta1,theta2) # 扁平化参数，25*401+10*26=10285
thetas.shape
```

```python
def deserialize(seq):  #反序列化
    return seq[:25*401].reshape(25,401) , seq[25*401:].reshape(10,26)
```

## 1.3 Feedforward and cost function 前馈和代价函数
### 1.3.1 Feedforward

     确保每层的单元数，注意输出时加一个偏置单元，s(1)=400+1，s(2)=25+1，s(3)=10。

```python
def simgoed(z):
    return 1/(1+np.exp(-z))
```

```python
def Feedforward(theta,x):
    t1,t2 = deserialize(theta)
    a1=x
    z2 = np.dot(a1,t1.T)
    a2 = np.insert(simgoed(z2),0,1,axis=1)
    z3 = np.dot(a2,t2.T)
    a3 = simgoed(z3)
    return a1,z2,a2,z3,a3
```

```python
a1,z2,a2,z3,h = Feedforward(thetas,x)
```

# 1.3.2 Cost function

### 回顾下神经网络的代价函数（不带正则化项）

### 输出层输出的是对样本的预测，包含5000个数据，每个数据对应了一个包含10个元素的向量，代表了结果有10类。在公式中，每个元素与log项对应相乘。

### 最后我们使用提供训练好的参数θ，算出的cost应该为0.287629

```python
def computeCost(thetas,x,y):
    a1,z2,a2,z3,h = Feedforward(thetas,x)
    first = -y*np.log(h)
    second = -(1-y)*np.log(1-h)
    J = first + second
    return J.sum()/len(x)
    
```

```python
computeCost(thetas,x,y)
```

### 1.4 Regularized cost function 正则化代价函数

#### 注意不要将每层的偏置项正则化。

#### 最后You should see that the cost is about 0.383770

```python
def Regularized_Cost(thetas,x,y,mylambda = 1):#mylambda是公式中的lamda
    #计算后面的正则项
    t1,t2= deserialize(thetas)
    #正则是不对每一层的偏置，即第一列进行正则的
    reg = np.sum(t1[:,1:]**2) + np.sum(t2[:,1:]**2)
    return computeCost(thetas,x,y)+mylambda / (2 * len(x)) * reg 
```

```python
Regularized_Cost(thetas,x,y)
```

## 2 Backpropagation 反向传播
### 2.1 Sigmoid gradient S函数导数
        这里可以手动推导，并不难。

```python
def sigmoid_gradient(z):
    return simgoed(z) * (1-simgoed(z))
```

## 2.2 Random initialization 随机初始化

#### 当我们训练神经网络时，随机初始化参数是很重要的，可以打破数据的对称性。一个有效的策略是在均匀分布(−e，e)中随机选择值，我们可以选择 e = 0.12 这个范围的值来确保参数足够小，使得训练更有效率。

```python
def random_initialization(size):#在这个范围内取出数目的数量
    return np.random.uniform(-0.12,0.12,size)
```

## 2.3 Backpropagation 反向传播
#### 目标：获取整个网络代价函数的梯度。以便在优化算法中求解。
#### 这里面一定要理解正向传播和反向传播的过程，才能弄清楚各种参数在网络中的维度，切记。比如手写出每次传播的式子。

```python
print("a1",a1.shape,"　theta1",theta1.shape)
print("z2",z2.shape)
print("a2",a2.shape,"　theta2",theta2.shape)
print("z3",z3.shape," h",h.shape)
```

<img src="./backfeedforward.png" />

```python
def gradient(thetas,x,y):
    t1,t2 = deserialize(thetas)
    a1,z2,a2,z3,h = Feedforward(thetas,x)
    d3 = h - y#先是预测的输出值减去实际的值  (5000,10)
    d2 = np.dot(d3,theta2[:,1:])*sigmoid_gradient(z2)  #这里也不考虑第一列的偏置  (5000,25)
    D2 = np.dot(d3.T,a2)  #(10,26)这里函数的偏导数
    D1 = np.dot(d2.T,a1)  #(25,401)
    D = 1/(len(x))*serialize(D1,D2)
    return D
```

## 2.5 regularized_gradient 正则化神经网络


```python
def regularized_gradient(theta, X, y, l=1): 
    """不惩罚偏置单元的参数""" 
    a1, z2, a2, z3, h = Feedforward(theta, X) 
    D1, D2 = deserialize(gradient(theta, X, y)) 
    theta1[:,0] = 0 
    theta2[:,0] = 0 
    reg_D1 = D1 + (l / len(X)) * theta1 
    reg_D2 = D2 + (l / len(X)) * theta2 
    return serialize(reg_D1, reg_D2)
```

## 2.5 Gradient checking 梯度检测

#### 在你的神经网络,你是最小化代价函数J(Θ)。执行梯度检查你的参数,你可以想象展开参数Θ(1)Θ(2)成一个长向量θ。通过这样做,你能使用以下梯度检查过程。

```python
def numerical_grandient_check(thetas,x,y,e):
    def numerical_grandient(plus,minus,e):
        """
            用于计算数值上的梯度，更好的评估通过偏导数得到的梯度是否是合理的
        """
        return (Regularized_Cost(plus,x,y) - Regularized_Cost(minus,x,y))/(2*e)
    
    numrical_grds = []
    for i in range(len(thetas)):
        plus = thetas.copy() #拷贝其值，将其加上　减去　e
        minus = thetas.copy()
        plus +=e
        minus +=e
        grad_i = numerical_grandient(plus,minus,e)
        numrical_grds.append(grad_i)
    numrical_grds = np.array(numrical_grds)
    analytic_grds = regularized_gradient(thetas,x,y)
    diff = np.linalg.norm(numrical_grds - analytic_grds) /np.linalg.norm(numrical_grds + analytic_grds)
    print("the relative difference will be smaller than 10e-9(epsilon=0.0001).\nDifference: {}\n".format(diff))
```

```python
numerical_grandient_check(thetas,x,y,0.0001) #epsilon = 0.0001
```

## 2.6 Learning parameters using fmincg 优化参数

```python
def trainNN(x,y):
    init_x = random_initialization(10285)  # 25*401 + 10*26
    res = scipy.optimize.minimize(fun=Regularized_Cost,
                                  x0 = init_x,
                                  args = (x,y,1),
                                  method = "TNC",
                                  jac = regularized_gradient,
                                  options = {"maxiter":4000})
    return res
```

```python
res = trainNN(x,y)
res
```
